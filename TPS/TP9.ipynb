#### Utiliser les méthodes usuelles de pandas pour obtenir les informations de base sur ce dataset (nombre d'observations et de variables, types des variables, valeurs manquantes, etc.).

```python
import pandas as pd

df = pd.read_csv('WHR-2019.csv')

info = df.info()
head = df.head()
missing_values = df.isnull().sum()

info, head, missing_values
```

#### Représenter chacune des 9 variables quantitatives par un histogramme. On disposera ceux-ci dans une grille virtuelle de 3 lignes et 3 colonnes. On pourra utiliser la fonction 'subplots' de Matplotlib pour construire la grille puis seaborn pour créer les histogrammes.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Variables quantitatives à représenter
quantitative_vars = ['Life Ladder', 'Log GDP per capita', 'Social support', 
                     'Healthy life expectancy at birth', 'Freedom to make life choices', 
                     'Generosity', 'Perceptions of corruption', 'Positive affect', 
                     'Negative affect']

#histogramme pour chaque variable
plt.figure(figsize=(15, 15))
for i, var in enumerate(quantitative_vars, 1):
    plt.subplot(3, 3, i)
    sns.histplot(df[var], kde=True)
    plt.title(var)

plt.show()
```

#### Même question avec des courbes de densité puis avec des boîtes à moustache.

```python
import matplotlib.pyplot as plt
import seaborn as sns

#les courbes de densité pour chaque variable
plt.figure(figsize=(15, 15))
for i, var in enumerate(quantitative_vars, 1):
    plt.subplot(3, 3, i)
    sns.kdeplot(df[var], fill=True)
    plt.title(var)

plt.show()
```

```python
# Créer les boîtes à moustache pour chaque variable
plt.figure(figsize=(15, 15))
for i, var in enumerate(quantitative_vars, 1):
    plt.subplot(3, 3, i)
    sns.boxplot(data=df[var])
    plt.title(var)

plt.show()
```

#### Calculer les paramètres de forme de ces variables.

```python
# Calcule de l'asymétrie 
skewness = df.skew()

# Calcule de l'aplatissement 
kurtosis = df.kurtosis()

# Afficher les paramètres de forme
print("Asymétrie (Skewness) :")
print(skewness)
print("\nAplatissement (Kurtosis) :")
print(kurtosis)
```
